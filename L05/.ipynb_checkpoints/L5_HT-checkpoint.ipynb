Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '123', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['123'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
def input_comand(): #ввод команды
    result = input("Введите команду\n")
    return result

def comand_id(command): #идентификация команды
    if command == 'p':
        p_command()
    elif command == 's':
        s_command()
    elif command == 'l':
        l_command()
    elif command == 'as':
        as_command()
    elif command == 'd':
        d_command()
    elif command == 'm':
        m_command()
    elif command == 'a':
        a_command()
    elif command == 'q':
        print("Удачного дня")
        
def find_doc(what, where):
    f=0
    for element in documents:
        if what == element[where]:
            f=1
    return f

def find_direct(direct_if):
    f=0
    for element in directories:
        if direct_if == element:
            f=1
    return f

def find_doc_num():
    f=0
    while f == 0:
        doc_id = input("Введите номер документа\n")
        if find_doc(doc_id, 'number') == 0:
            print('Нет такого номера документа')
            f=0 
        else:
            f=1
    return doc_id


def p_command(): # people
    doc_id = find_doc_num()
    name = ''
    for data in documents:
        if doc_id == data['number']:
            name = data['name']
            break
    print("Имя владельца\n",name)
    comand_id(input_comand())

def s_command():
    doc_id = find_doc_num()
    direct =''
    for data in directories:
        if doc_id in directories[data]:
            direct = data
            break
    print("Номер полки\n", direct)
    comand_id(input_comand())   
    
def l_command():
    for data in documents:
         print(data['type'] +' "' + data['number'] +'" "' + data['name'] +'"')
    comand_id(input_comand())
            
def as_command():
    f=0
    while f == 0:
        new_direct = input("Введите номер новой полки\n")
        if find_direct(new_direct) ==1:
            print('Такая полка уже есть')
            f=0 
        else:
            f=1        
    directories[new_direct] = []
    print(f'Новая полка { new_direct } успешно создана')
    comand_id(input_comand())

def d_command():
    doc_id = find_doc_num()
    new_documents = []
    global documents
    for element in documents:
        if element['number'] != doc_id:
            new_documents.append(element)
    documents = new_documents
    
    direct =''
    for data in directories:
        if doc_id in directories[data]:
            direct = data
            break
    
    direct_number_list = []
    for data in directories[direct]:
        if data != doc_id:
            direct_number_list.append(data)
    directories[direct] = direct_number_list
    
    print(f'Документ с номером { doc_id } успешно удален из каталога и { direct } полки')
    comand_id(input_comand())

def m_command():
    doc_id = find_doc_num()
    f=0
    while f == 0:
        new_direct = input("Введите номер целевой полки\n")
        if find_direct(new_direct) != 1:
            print('Такой полки не существует')
            f=0 
        else:
            f=1
    direct =''
    for data in directories:
        if doc_id in directories[data]:
            direct = data
            break
    
    direct_number_list = []
    for data in directories[direct]:
        if data != doc_id:
            direct_number_list.append(data)
    directories[direct] = direct_number_list
    directories[new_direct].append(doc_id)
    print(f'Документ с номером { doc_id } успешно перемещен с полки { direct } на полку { new_direct }')
    comand_id(input_comand())

def a_command():
    new_doc={}

    f=0
    while f == 0:
        new_doc['number'] = input("Введите номер документа\n")
        if find_doc(new_doc['number'], 'number') != 0:
            print('Такой номер документа уже существует')
            f=0 
        else:
            f=1
    
    f=0
    while f == 0:
        new_direct = input("Введите номер целевой полки\n")
        if find_direct(new_direct) != 1:
            print('Такой полки не существует')
            f=0 
        else:
            f=1

    new_doc['type'] = input("Введите тип документа\n")
    new_doc['name'] = input("Введите имя владельца\n")
    
    documents.append(new_doc)
    directories[new_direct].append(new_doc['number'])
    new_doc_for_print = new_doc['number']
    print(f'Документ с номером { new_doc_for_print } успешно добавлен в каталог и на полку { new_direct }')
    
    comand_id(input_comand())
```

```python
def main():
    """
Перечень доступных команд

p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится;
l – list – команда, которая выведет список всех документов;
as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень;
d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок;
m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую;
a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.
q - quite - покинуть редактор  
    """
    comand_id(input_comand())
```

```python
main()
```
